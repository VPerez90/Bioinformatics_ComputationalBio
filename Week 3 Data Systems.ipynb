Python code for calculating substitution matrices

```python
# Need to import log function first from math
from math import log

# define function calculateSubMatrix for calculating substitution matrices
def calculateSubMatrix(alignment, alphabet, maxValue, smoothData=5):

  subMatrix = {} # This empty dict will contain substitution matrix
  counts = {} # Dict containing residue counts
  
  # Will measure number of residues and number of substitutions of one residue to another
  for letterA in alphabet:
    subDict = {}
    
    for letterB in alphabet:
      subDict[letterB] = 0
  
    subMatrix[letterA] = subDict
    counts[letterA] = 0
  
  # Will need these for calculations later on - 0.0 because division will occur
  totalResidue = 0.0
  totalSub = 0.0
  
  # Most important part of the function:
  # Loops through all input alignments and collects substitutional data
  for align in alignment:
 
    numberPosition = len(align[0]) #saves the length of the sequence in variable - 0 is first sequence

    for i in range(numberPosition):
      
      # Initialize list for collecting residues 
      letters = []
      
      for sequence in align:
        
        # If gap, skip and move on
        letter = sequence[i]
        if letter == '-':
          continue
        
        # Add resdiue to list
        letters.append(letter)

      # Once residues are collected, go throught again to get possible pairs 
      for letterA in letters:
        counts[letterA] += 1 # Count for each letter increases by one
      
        for letterB in letters:          
          subMatrix[letterA][letterB] += 1 # Add to matrix

      numLetters = len(letters) # length so far, not including gaps
      totalResidue += numLetters # Add length of letters to totalResidue   
      totalSub += numLetters * numLetters # Substitution total increased by numLetters^2

  # Blank dict for saving average composition calculations 
  avgComposition = {}    
  
  # Calculate average composition   
  for letter in alphabet:
    avgComposition[letter] = counts[letter]/totalResidue      

  # Initialize variable maximumScore  
  maximumScore = None
  for residueA in alphabet:
    for residueB in alphabet:

      expected = avgComposition[residueA] * avgComposition[residueB]
      
      if not expected: # If calculation is 0, skip
        continue
      
      # pair count of combination 
      observed = subMatrix[residueA][residueB]
      weight = 1.0 / (1.0+(observed/smoothData)) 

      # Converted into probability
      observed /= totalSub
      observed = weight*expected + (1-weight)*observed #Apply weight calculated from above

      # log-odds scores is calculated by taking log of probability ratio 
      logOdds = log(observed/expected)
      
      # If a new maxScore is found, we will add score to the matrix
      if (maximumScore is None) or (logOdds>maximumScore):
        maximumScore = logOdds
      
      subMatrix[residueA][residueB] = logOdds
  
  # find absolute value of maxscore
  maxScore = abs(maximumScore)
 
  #Scale the log-odds values in substitution matrix and convert them to integers
  for resA in alphabet:
    for resB in alphabet:
      subMatrix[resA][resB] = int(maxValue*subMatrix[resA][resB]/maximumScore)

  return subMatrix


#Test function

randMat = ['AGTACGTATCGATGGACCTTA',
           'ATCCGTATGCCCTTAATTAGA',
           'ATGCGTGCAATGCTATTCCTA',
           'TTGCTATTAATGCCCTCCAAA',
           'ATCGTAGCTAGACTTTAGTAA',
           'ATCGTAGTCTGATTTTAACCA', 
           'ATCGTAGTCTGATCCCAATTA']


DNA = {'G': { 'G': 8, 'C':-5, 'A':-5, 'T':-5, 'N':0 },
       'C': { 'G':-5, 'C': 8, 'A':-5, 'T':-5, 'N':0 },
       'A': { 'G':-5, 'C':-5, 'A': 8, 'T':-5, 'N':0 },
       'T': { 'G':-5, 'C':-5, 'A':-5, 'T': 8, 'N':0 },
       'N': { 'G': 0, 'C': 0, 'A': 0, 'T': 0, 'N':0 }} 

nucleotides = DNA.keys()

print(calculateSubMatrix([randMat,], nucleotides, 10))
print('')
print('COOL!')
```

```python
# PUT IT ALL TOGETHER

from math import exp

seqData = ['AGTACGTATCGATGGACCTTA',
           'ATCCGTATGCCCTTAATTAGA',
           'ATGCGTGCAATGCTATTCCTA',
           'TTGCTATTAATGCCCTCCAAA',
           'ATCGTAGCTAGACTTTAGTAA',
           'ATCGTAGTCTGATTTTAACCA', 
           'ATCGTAGTCTGATCCCAATTA']

DNA = {'G': { 'G': 8, 'C':-5, 'A':-5, 'T':-5, 'N':0 },
       'C': { 'G':-5, 'C': 8, 'A':-5, 'T':-5, 'N':0 },
       'A': { 'G':-5, 'C':-5, 'A': 8, 'T':-5, 'N':0 },
       'T': { 'G':-5, 'C':-5, 'A':-5, 'T': 8, 'N':0 },
       'N': { 'G': 0, 'C': 0, 'A': 0, 'T': 0, 'N':0 }} 

def seqAlign(seqA, seqB, simMatrix=DNA, insert=8, extend=4): 
 

  """Variables numI and numJ represent the length of the rows and columns of the comparison grid â†’ 
     needs to be one length longer than each sequence"""
  numI = len(seqA) + 1
  numJ = len(seqB) + 1

  # Create two matrices. One for holding the scores and one for holding the route taken
  scoreMatrix = [[0] * numJ for x in range(numI)]
  routeMatrix = [[0] * numJ for x in range(numI)]
  
  # Make leftmost column have score start with 0 and decrease by 3 for each iteration
  for i in range(1, numI):
    routeMatrix[i][0] = 1
   
    
  # Make topmost row have score start with 0 and decrease by 3 for each iteration
  for j in range(1, numJ):
    routeMatrix[0][j] = 2
  
 # Nested loop used to fill in rest of the matrix 
  for i in range(1, numI):
    for j in range(1, numJ):
      
      # Penalty of a gap is 3 like our example from class 
      penalty1 = insert
      penalty2 = insert
      
      #NOT SURE HOW TO FIX THIS THOUGH TO MATCH THE 0,-3, -6â€¦.THIS IS WHAT I CAME UP WITH
      # Check to see if previous is a gap
      if routeMatrix[i-1][j] == 1:
        penalty1 = extend
        
      elif routeMatrix[i][j-1] == 2:
        penalty2 = extend
       
       # Look up similarity score of bases for seqA and seqB 
      similarity = simMatrix[seqA[i-1] ][seqB[j-1] ]
      
      # Make a list of three possible scores (left-up diagonal, up, and left)
      paths = [scoreMatrix[i-1][j-1] + similarity, # Route 0
               scoreMatrix[i-1][j] - penalty1, # Route 1
               scoreMatrix[i][j-1] - penalty2] # Route 2                     
      
      # The best of the three scores are assigned in â€˜bestâ€™ and the index of best score is assigned to route
      best = max(paths)
      route = paths.index(best)           

      # The scores and indexes for each iteration are saved in the matrix 
      scoreMatrix[i][j] = best
      routeMatrix[i][j] = route
   
   #  Initialize blank lists to hold the optimal sequence alignments for seqA and seqB    
  alignA = []
  alignB = []

  # Assigning variables to implement into while loop below for working backwards
  i = numI-1
  j = numJ-1
  score = scoreMatrix[i][j]

  while i > 0 or j > 0:
    route = routeMatrix[i][j]    

     # Determining which of the three routes had the highest score
    if route == 0: # Diagonal
      alignA.append( seqA[i-1] )
      alignB.append( seqB[j-1] )
      i -= 1
      j -= 1

    elif route == 1: # Gap in seqB
      alignA.append(seqA[i-1] )
      alignB.append('-')
      i -= 1      

    elif route == 2: # Gap in seqA
      alignA.append('-')
      alignB.append(seqB[j-1] ) 
      j -= 1
  
   # Reverse the alignment lists since we originally stored them backwards, now we want them forwards 
  alignA.reverse()
  alignB.reverse()
  alignA = ''.join(alignA)
  alignB = ''.join(alignB)

  return score, alignA, alignB 

def calcSeqSimilarity(seqA, seqB, simMatrix):

  numPlaces = min(len(seqA), len(seqB))
  
  totalScore = 0.0
  
  for i in range(numPlaces):
    
    residueA = seqA[i]
    residueB = seqB[i]
  
    totalScore += simMatrix[residueA][residueB]

  return totalScore

def getDistanceMatrix(seqs, simMatrix):

  numbers = len(seqs) #variable numbers for storing sequence lengths
  matrix = [[0.0] * numbers for x in range(numbers)]
  maxScores = [calcSeqSimilarity(x, x, simMatrix) for x in seqs]
  
  # Sequences are aligned using an input subsititution matrix
  for i in range(numbers-1):
    seqA = seqs[i]
  
    for j in range(i+1,numbers):
      seqB = seqs[j]
      
      # Alignment is passed back with a score which then gets converted into a distance
      score, alignA, alignB = seqAlign(seqA, seqB, simMatrix)
      maxScore = max(maxScores[i],maxScores[j])
      dist = maxScore - score
      
      matrix[i][j] = dist
      matrix[j][i] = dist

  return matrix

def getJoinPair(distMatrix):

  numbers = len(distMatrix) #variable for holding length of matrix
  
  minQ = None
  joinPair = None

  # nested loop for iterating through matrix  
  for i in range(numbers-1):
    sumRow = sum(distMatrix[i])
  
    for j in range(i+1, numbers):
      sumCol = sum(distMatrix[j])
    
      dist = distMatrix[i][j]
      q = (numbers-2)*dist - sumRow - sumCol # Most important here is how we calculate q
      
      if (minQ is None) or (q < minQ):
        minQ = q
        joinPair = [i,j] # If smallest so far, we save it as the joined pair aka closest sequences

  return joinPair


# Next function is to determine distance from a point on tree to the branch point...helps find distance to nodes
def getDistToJunction(distMatrix, i, j):
  
  numbers = len(distMatrix) #variable for holding length of matrix
  row = distMatrix[i] # i is row
  column = distMatrix[j] # j is column
 
  dist = distMatrix[i][j] + (sum(row)-sum(column))/(numbers-2) # calculate distance
  dist *= 0.5 #take that and divide by 2

  return dist

# Last function builds the tree
def neighborJoinTree(distMatrix):
 
  joinOrder = [] #empty list
  numbers = len(distMatrix) #variable for holding length of matrix
  tree = list(range(numbers))  # list needed for Python 3
  
  # Using a while loop to iterate while numbers is greater than 2
  while numbers > 2:

    x, y = getJoinPair(distMatrix)

    # New branch point is joined
    node = (tree[x], tree[y])
    joinOrder.append(node)
    tree.append(node)

    del tree[y]
    del tree[x]

    distX = getDistToJunction(distMatrix, x, y)
    distY = getDistToJunction(distMatrix, y, x)
  
    distMatrix.append([0] * (numbers+1))

    for i in range(numbers):
      if i not in (x,y):

        dist = (distMatrix[x][i]-distX) + (distMatrix[y][i]-distY)
        dist *= 0.5
  
        distMatrix[i].append(dist)
        distMatrix[numbers][i] = dist

    del distMatrix[y]
    del distMatrix[x]
  
    for row in distMatrix:
      del row[y]
      del row[x]

    numbers -= 1

  tree = tuple(tree) #Change tree to a tuple
  joinOrder.append(tree)
  
  return tree, joinOrder

#TEST TO SEE IF IT WORKS

distMatrix = getDistanceMatrix(seqData, DNA)

tree, treeJoiner = neighborJoinTree(distMatrix)

print('Here is the tree:')
print(tree)
print('')
print("I DID IT!!!!")
```
